# Data Visualization

#### Reasons for visualizing data :

* Observing the pattern.
* Identifying extreme values that could be anomolies.
* Easy interpretation

#### Popular plotting libraries in Python :

1. matplotlib - To create 2D graphs and plots

## Matplotlib
---

* Matplotlib is a 2d plotting library which produces good quality figures.
* Although it has origins in emulating the MATLAB graphics commands, it is independent of MATLAB.
* It makes heavy use of Numpy and other extension codes to provide good performance even for large arrays.

### Scatter Plots
---

* A scatter plot is a set of points that represent s the values obtained for two different variables plotted on a horizontal and vertical axes.

* Scatter plots are used to convey the relationship between two numerical variables.
* Scatterplots are sometimes called as "correlation plots" because they show how the two variables are correlated.

Let's import the necessary libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Let's import the dataset

```python
cars_data = pd.read_csv("F:\Google Drive (Dataman Udit)\Python Coaching\Data Science with Python\Pandas Datasets\Toyota.csv",
                       index_col = 0, na_values = ["??", "????"])
cars_data.head()
```

Let's remove the missing values from the dataset

```python
cars_data.dropna(axis = 0, inplace = True)
```

Let's plot the scatter plot

```python
plt.scatter (x ='Age', y = 'Price', data = cars_data, color = 'red')
plt.title('Scatter Plot of Price Vs. Age of the cars')
plt.xlabel('Age (Months)')
plt.ylabel('Price (Euros)')
plt.show
```

### Histogram
---


* It is a graphical representation of data using bars of different heights.
* Histogram groups numbers into ranges and the height of each bar depicts the frequency of each range or, bin.
* Histogram is used to represent the frequency distribution of the numerical variables.

```python
plt.hist(cars_data["KM"])
```

```python
plt.hist(cars_data["KM"],
        color = "green",
        edgecolor = "white",
        bins = 5)
plt.title("Histogram of Kilometers")
plt.xlabel("Kilometer")
plt.ylabel("Frequency")
plt.show()
```

The above frequency distribution of "kilometers" shows that most of the cars have travelled between 50000-100000 KM and there are only few cars with more distance travelled. 

### Bar Plot
---

* A bar plot is a plot that represents categorical data wih rectangular bars with lengths propotional to the counts that they represent.
* Bar plot is used to represent the frequency distribution of categorical variables.
* Its easy to compare sets of data between different groups using bar plots.

```python
counts = [979,120,59]
fueltype = ("Petrol", "Diesel", "CNG")
index = np.arange(len(fueltype))

plt.bar(index, counts, color = ["red", "blue", "cyan"])
plt.title("Bar plot of fuel type")
plt.xlabel("Fuel Types")
plt.ylabel("Frequency")
plt.show()
```

## SEABORN
---

* Seaborn is a Python data visualization library based on matplotlib.
* It provides a high level interface for drawing attractive and informative statistical graphics.

```python
#Importing Seaborn

import seaborn as sns
```

### Scatter Plot
---

```python
sns.set(style = "darkgrid") #Default style of seaborn
sns.regplot(x = "Age", y = "Price", data = cars_data)
```

* By default `fit_reg = True`
* It estimates and plots a regression model relating the X and Y variables.

```python
#Scatter plot without regression line

sns.regplot(x = cars_data["Age"], y = cars_data["Price"], fit_reg= False)
```

```python
#Changing the default marker to "*"

sns.regplot(x = cars_data["Age"], y = cars_data["Price"], fit_reg=False, marker="*")
```

* Using the `hue` parameter, including another variable to show the fuel types categories with different colors

```python
sns.lmplot(x = "Age", y = "Price", data = cars_data, fit_reg = False, hue = "FuelType", legend = True, palette = "Set1")
```

### Histogram
---

* Seaborn plots histogram with kernel density function by default.

```python
sns.distplot(cars_data["Age"])
```

To remove the kernel density function, we have to set `kde = False`

```python
sns.distplot(cars_data["Age"],kde =  False)
```

To set the desired bin size, we have to set `bins = xx`

```python
sns.distplot(cars_data["Age"], kde = False, bins = 5)
```

### Bar Plot
---

Frequency distribution of the fuel type of the cars.

```python
sns.countplot(cars_data["FuelType"])
```

### Clustered Bar Plot
---

Grouped bar plot of `FuelType` and `Automatic`

```python
sns.countplot(x = "FuelType", hue = "Automatic", data = cars_data)
```

### Box & Whiskers Plot - Numerical Variable
---

Box and whiskers plot of `Price` to visually interpret the five-summary statistics. 

```python
sns.boxplot(y = cars_data["Price"])
```

To show the plot horizontally, we have to set `x = cars_data["Price"]` instead of `y = cars_data["Price"]`

```python
sns.boxplot(x = cars_data["Price"])
```

### Box & Whiskers Plot for Numerical & Categorical Variables
---

* Price of the cars for various fuel types -:

```python
sns.boxplot(x = cars_data["FuelType"], y = cars_data["Price"])
```

### Clustered Box & Whiskers Plot
---

Grouped box and whiskers plot of `Price` Vs. `FuelType` and `Automatic` -:

```python
sns.boxplot(x = "FuelType", y = "Price", hue = "Automatic", data = cars_data)
```

### Box & Whiskers Plot with Histogram
---

To plot two charts in the same window, first we have to split the charting area into 2 parts with desired proportions.

```python
f,(ax_box, ax_hist) = plt.subplots(2, gridspec_kw = {"height_ratios" : (.15,.85)})
```

```python
f,(ax_box, ax_hist) = plt.subplots(2, gridspec_kw = {"height_ratios" : (.15,.85)})
sns.boxplot(cars_data["Price"], ax = ax_box)
sns.distplot(cars_data["Price"], ax = ax_hist, kde = False)
```

### Pairwise Plot
---

* It iis used to plot pairwise relationship in a dataset.
* It creates scatter plot for joint relationships and density curves for univariate distributions.

```python
sns.pairplot(cars_data, kind = "scatter", hue = "FuelType")
plt.show()
```
